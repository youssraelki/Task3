```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset from the CSV file into a DataFrame
df = pd.read_csv('C:/Users/user/Downloads/train.csv')

# Display the first few rows of the DataFrame to ensure it's read correctly
print(df.head())

# Select a specific column to visualize
Age = 'Age'  # Replace with your column name

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(df[Age].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title(f'Histogram of {Age}')
plt.xlabel(Age)
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Create a bar chart for categorical data
if 'Sex' in df.columns:
    plt.figure(figsize=(10, 6))
    df['Sex'].value_counts().plot(kind='bar', edgecolor='k', alpha=0.7)
    plt.title('Bar Chart of Sex')
    plt.xlabel('Sex')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()
```
